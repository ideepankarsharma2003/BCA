```python
# import libraries
import pandas as pd
import numpy as np
```

```python
## data Cleaning 
# dataset is stored in varialbe data
data  = pd.read_csv("austin_weather.csv")
```

```python
# print the data
data
```

```python
# droping the unnessery values/columns
data = data.drop(['Events','Date','SeaLevelPressureAvgInches','SeaLevelPressureLowInches','SeaLevelPressureHighInches'],axis=1)
```

```python
# removing the unNumerical values 
data = data.replace('T',0.0)
data = data.replace('-',0.0)
```

```python
# data is now preproccsed , finally save the data 
data.to_csv("austin_final.csv")
```

```python
## start of next step that is linear model
# import libraries for the model

import sklearn as sk
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
## read the cleanes data
data  = pd.read_csv("austin_final.csv")
```

```python
data.columns
```

```python
# store all the feature/indem=pendent variable to x and lables to y
x = data.drop(['PrecipitationSumInches'],axis=1)
# y will serve as labels
y = data['PrecipitationSumInches']
y = y.values.reshape(-1,1)
```

```python
x
```

```python
y
```

```python
# observe a random day by plotting a graph
day_index = 776
days = [i for i in range (y.size)]
```

```python
# reshaping the values of y into 1d column matrix
```

```python
# initilize the Linear Regression Classifier
clf = LinearRegression()
```

```python
#Train the classifir with an input data
clf.fit(x,y)
```

```python
# give a sample input to test our model
inp = np.array([[74, 60, 45, 67, 49, 43, 33, 45,57, 29.68, 10, 2, 0, 20, 4, 31]])
# inp = inp.reshape(1, -1)
inp
```
